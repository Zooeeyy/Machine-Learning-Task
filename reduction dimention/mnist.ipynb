```python
import tensorflow as tf
from sklearn.decomposition import PCA
```

```python
train, test = tf.keras.datasets.mnist.load_data(path='mnist.npz')
Xtrain, ytrain = train
Xtest, ytest = test
```

```python
Xtrain = Xtrain[0:5000]
ytrain = ytrain[0:5000]
Xtrain = Xtrain/255

Xtrain = Xtrain.reshape(5000, 784)
```

# PCA

```python
from sklearn import decomposition

# Membuat objek PCA
pca = decomposition.PCA(n_components=2)

# Melakukan PCA pada data
pca_data = pca.fit_transform(Xtrain)

# Cetak hasil PCA
print(pca_data)
```

```python
import plotly.express as px
fig = px.scatter(x=pca_data[:,0], y=pca_data[:,1], color=ytrain)
fig.show()
```
